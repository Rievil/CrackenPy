```python
from crackest.cracks import CrackPy

cp=CrackPy(model=0)
```

```python
cp.get_mask(impath=r"../Img/ID14_940_Image.png")
cp.distancemap()
```

```python
cp.show_mask(mask="crack")
```

```python
cp.overlay(figsize=(12,3))
```

```python
cp.save(r"../Plots/Example.png")
```

```python
cp.get_metrics()
```

```python
from crackest.crack_analyzer import CrackAn

ca=CrackAn(cp=cp)
ca.input(file=r"../Img/256_Image_29-01-2022 02-36-02.png")
ca.registr()
```

```python
ca.preview()
ca.fig.savefig(r"../Plots/Multiple_registr.png")
```
